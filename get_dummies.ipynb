```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df= pd.read_csv('diabetes_risk_prediction_dataset.csv')
df.head()
```

```python
cat= ['Gender',	'Polyuria',	'Polydipsia',	'sudden weight loss',	'weakness',	'Polyphagia',	'Genital thrush',	'visual blurring',	'Itching', 'Irritability',	'delayed healing', 	'partial paresis',	'muscle stiffness', 'Alopecia',	'Obesity'] #kategorijske promenljive

df= pd.get_dummies(df,columns= cat, drop_first = False)
df.head()
```

```python
y=df['class']
X= df.drop(columns=['class'])
X.head()
print(X.shape)
y= pd.get_dummies(y, drop_first= True) #dummies za sao kada je class pozitivan tj ima dijabetes
y.head()

Xtrain, Xtest, ytrain, ytest= train_test_split(X, y, test_size= 0.2)
print(Xtrain.shape, ytrain.shape, Xtest.shape, ytest.shape)
```

```python

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report

ks= [x for x in range (1, 2, 1)]

models= [KNeighborsClassifier(n_neighbors= k)for k in ks]

for model in models:
  print(f"k= {model.get_params()['n_neighbors']}")
  model.fit(Xtrain, ytrain)
  y_pred= model.predict(Xtest)
  print(classification_report(ytest, y_pred))
```
