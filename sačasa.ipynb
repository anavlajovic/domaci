```python
from sklearn.datasets import make_classification
import seaborn as sns
X, y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, n_repeated=0, n_classes=3, n_clusters_per_class=1)
import pandas as pd

data= pd.DataFrame(X, columns=['x1', 'x2'])
data= pd.concat([data, pd.DataFrame(y, columns=['y'])], axis=1)
data.head()

from sklearn.model_selection import train_test_split
train, test= train_test_split(data, test_size=0.2)

import seaborn as sns
sns.scatterplot(train, x='x1', y='x2', hue='y')
```
